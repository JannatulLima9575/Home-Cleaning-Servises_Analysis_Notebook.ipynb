## Selecting Rows and Columns from DataFrame

There are two major ways to do it. One is by selecting the rows and columns numerically by using df.iloc() method. The other one is to use df.loc to select the columns by labels

## df.iloc

```python
# selecting the 125th row
df.iloc[124]
```

```python
# selecting the first 10 rows
df.iloc[:5]
```

```python
# selecting the 4th column count
df.iloc[:, 5]
```

```python
df.iloc[:, 2:4]
```

```python
# selecting specific rows and columns, 2 to 4th row and 2 to 3rd column
df.iloc[2:5, 2:4]
```

```python
# selecting a single element
df.iloc[2 , 2]
```

## df.loc

```python
# selecting the ethnicity column
df.loc[:, 'Ethnicity']
```

```python
# selecting column year of birth, gender and ethnicity for rows 500 - 700

df.loc[500:700, ['Year of Birth', 'Gender', 'Ethnicity']]
```
