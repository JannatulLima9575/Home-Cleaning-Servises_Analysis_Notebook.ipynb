```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
```

```python
data = {
    'Name': ['Lima', 'Shima', 'Jannat', 'Ema', 'Idrish','Musa', 'Easa', 'Happy', 'Sanny', 'Popy','Riha', 'Bob', 'Charlie', 'Nura', 'Nuha','Alice', 'Ruba', 'Rubi', 'Runi', 'Eva'],
    'Subject': ['Math', 'English', 'Science', 'Math', 'English', 'Science','Math', 'English', 'Science','Math', 'English', 'Science','Math', 'English', 'Science','Math', 'English', 'Science','Math', 'English'],
    'Marks': [99, 78, 92, 88, 95,85, 78, 92, 88, 95,85, 78, 92, 88, 95,85, 78, 92, 88, 95],
    'Semester': ['8th', '8th', '8th', '8th','8th', '8th', '8th', '8th', '8th','8th', '8th', '8th', '8th', '8th','8th', '8th', '8th', '8th', '8th','8th']
}

df = pd.DataFrame(data)

print("Student Dataset:")
print(df)
```

```python
average_marks = df['Marks'].mean()
print("\nAverage Marks:", average_marks)

top_scorer = df[df['Marks'] == df['Marks'].max()]
print("\nTop Scorer:")
print(top_scorer)
```

```python
high_scorers = df[df['Marks'] > 80]
print("\nStudents who scored more than 80:")
print(high_scorers)
```

```python
while True:
    print("\n--- Add a New Student ---")
    name = input("Enter Student Name (or type 'exit' to stop): ")
    if name.lower() == 'exit':
        break

    subject = input("Enter Subject: ")
    marks = int(input("Enter Marks: "))
    semester = input("Enter Semester: ")

    new_row = pd.DataFrame({'Name': [name], 'Subject': [subject], 'Marks': [marks], 'Semester': [semester]})
    df = pd.concat([df, new_row], ignore_index=True)

    print("\nUpdated Dataset:")
    print(df)
```

```python
file_name = "student_data.csv"

df.to_csv(file_name, index=False)
print(f"\nDataset has been saved to {file_name}.")

print("\n--- Loading Data from CSV File ---")
loaded_df = pd.read_csv(file_name)
print("Loaded Dataset:")
print(loaded_df)
```

```python
def assign_grade(marks):
    if marks >= 90:
        return 'A+'
    elif marks >= 80:
        return 'A'
    elif marks >= 70:
        return 'B'
    else:
        return 'C'

df['Grade'] = df['Marks'].apply(assign_grade)
print(df)
```

```python
def assign_grade(marks):
    if marks >= 90:
        return 'A+'
    elif marks >= 80:
        return 'A'
    elif marks >= 70:
        return 'B'
    else:
        return 'C'

df['Grade'] = df['Marks'].apply(assign_grade)

grade_counts = df['Grade'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightblue', 'lightgreen', 'salmon'])
plt.title('Grade Distribution')
plt.show()
```

```python
average_marks_by_semester = df.groupby('Semester')['Marks'].mean()

plt.figure(figsize=(10, 6))
plt.plot(average_marks_by_semester.index, average_marks_by_semester.values, marker='o', linestyle='-', color='b')
plt.title('Average Marks by Semester', fontsize=16)
plt.xlabel('Semester', fontsize=12)
plt.ylabel('Average Marks', fontsize=12)
plt.grid(True)
plt.show()
```
